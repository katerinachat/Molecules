Merging the two dataframes only captures 47 out of 147 authors (30 out of 90 for subset), likely because of different formattings in the 'digital first' dataframe. Not all entries are strictly "Last Name, First Name" for example, we also have "Last Name, First Initial". Let's try fuzzy matching..

https://towardsdatascience.com/how-to-do-fuzzy-matching-in-python-pandas-dataframe-6ce3025834a6
https://medium.com/@rtjeannier/combining-data-sets-with-fuzzy-matching-17efcb510ab2

```python
from fuzzywuzzy import fuzz

def match_name(name, list_names, min_score=0):
    # -1 score incase we don't get any matches
    max_score = -1
    # Returning empty name for no match as well
    max_name = ""
    # Iternating over all names in the other
    for name2 in list_names:
        #Finding fuzzy match score
        score = fuzz.ratio(name, name2)
        # Checking if we are above our threshold and have a better score
        if (score > min_score) & (score > max_score):
            max_name = name2
            max_score = score
    return (max_name, max_score)
```

```python
# List for dicts for easy dataframe creation
dict_list = []
# iterating over our players without salaries found above
for name in ceidr_df['Name']:
    # Use our method to find best match, we can set a threshold here
    match = match_name(name, digfirst_doi['Name'].unique(), 75)
    
    # New dict for storing data
    dict_ = {}
    dict_.update({"Name" : name})
    dict_.update({"Matching_Name" : match[0]})
    dict_.update({"Score" : match[1]})
    dict_list.append(dict_)
    
merge_table = pd.DataFrame(dict_list)
# Display results
merge_table = merge_table[merge_table['Score'] > 0]
print(merge_table.shape)
print(merge_table.to_string())
```

Some names don't match well, "Darlow, Christopher" : "Brown, Christopher" (index 60) and "Hall, Jamie" : "Hall, James" (index 43)

```python
merge_table = merge_table.drop([43,60], axis=0)
print(merge_table.shape)
print(merge_table.to_string())
```

```python
corrected_ceidr = pd.merge(ceidr_df, merge_table, on=['Name'])
corrected_ceidr = corrected_ceidr.drop(['Name', 'Score'], axis=1)
corrected_ceidr.shape
```

```python
ceidr_digfirst = pd.merge(digfirst_doi, corrected_ceidr, left_on='Name', right_on='Matching_Name')
print(ceidr_digfirst.shape)
print('There are {} unique entries'.format(len(ceidr_digfirst['Name'].unique())))
```
