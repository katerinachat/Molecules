```python
import os
import pandas as pd
```

```python
digfirst_df = pd.read_csv('../download_papers/data/dig_first.csv', names=['Department', 'Name', 'Staff', 'Authors', 'Title', 'DOI'], header=0)
digfirst_df = digfirst_df[['Name', 'DOI']].dropna()
digfirst_df.head()
```

```python
digfirst_df = digfirst_df.groupby(['DOI'])['Name'].apply(lambda x: ';'.join(x)).reset_index()
digfirst_df.tail()
```

```python
doi_map_df = []
for filename in os.listdir('../download_papers/logs'):
    if 'doi_map.csv' in filename:
        df = pd.read_csv('../download_papers/logs/{}'.format(filename), names=['DOI', 'identifier'], header=0)
        doi_map_df.append(df)
    else:
        continue
len(doi_map_df)
```

```python
doi_df = pd.concat(doi_map_df, axis=0, ignore_index=True)
doi_df
```

```python
merged_df = pd.merge(digfirst_df, doi_df, on="DOI")
merged_df
```

```python
merged_df.to_csv(r'doi_author_map.csv', index = False)
```
